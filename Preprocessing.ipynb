#### ***Fill the Age null values***

```python
titanic_data['Age'] = titanic_data['Age'].fillna(29)
```

```python
titanic_data.describe(include='all')
```

### ***OneHotEncoding of Embarked and Sex Columns***

```python
titanic_data['Embarked'] = titanic_data['Embarked'].fillna('Q')
```

```python
titanic_data.info()
```

```python
dummies = pd.get_dummies(titanic_data['Embarked'])
dummies
```

```python
titanic_data['Sex'] = titanic_data['Sex'].map({'male':0, 'female':1})
```

```python
merged = pd.concat([titanic_data, dummies], axis=1)
merged
```

```python
columns_to_drop = ['Embarked', 'Name', 'Ticket', 'Cabin']

dropped_columns = merged.drop(columns_to_drop, axis=1)
```

```python
dropped_columns.head()
```

```python
dropped_columns.info()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)
for train_indices, test_indices in split.split(dropped_columns, dropped_columns[['Survived', 'Pclass', 'Sex']]):
    titanic_train_set = dropped_columns.loc[train_indices]
    titanic_test_set = dropped_columns.loc[test_indices]
```

#### *Check if the data is evenly distributed*

```python
plt.subplot(1,2,1)
titanic_train_set['Survived'].hist()
titanic_train_set['Pclass'].hist()
#titanic_train_set['Sex'].hist()

plt.subplot(1,2,2)
titanic_test_set['Survived'].hist()
titanic_test_set['Pclass'].hist()
#titanic_test_set['Sex'].hist()

plt.show()
```

```python
titanic_train_set.head()
```

#### ***Scale the Inputs***

```python
from sklearn import preprocessing
```

```python
unscaled_train_input = titanic_train_set.drop('Survived', axis=1)
unscaled_train_target = titanic_train_set['Survived']

unscaled_test_input = titanic_test_set.drop('Survived', axis=1)
unscaled_test_target = titanic_test_set['Survived']
```

```python
scaled_train_input = preprocessing.scale(unscaled_train_input)
scaled_test_input = preprocessing.scale(unscaled_test_input)
```

#### ***Save the data in .npz files***

```python
np.savez('titanic_data_train', inputs=scaled_train_input, targets=unscaled_train_target)
np.savez('titanic_data_test', inputs=scaled_test_input, targets=unscaled_test_target)
```
